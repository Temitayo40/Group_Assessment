we can also compute the correlation coefficient btw the columns using .corr
correlation can be interpreted in the fgg ways;

Strenght: The greater the absolute value of the correlation coefficient, the stronger the relatioship. The extreme values of -1 and 1 indicates a perfect linear relationship where  change in one variable is caused by chane in the other.
A coefficient of zero represents no linear relationship.
    when value is in-between 0 and +1/-1, there is a relatioship but the points don't all fall in line.
    
Direction: Positive coefficients indicate that when the value of one variable increases, the value of the other variable also tends to increase.
while negative coefficients represent cases when value of one variable increases, the value of the other variables tend to decrease.
